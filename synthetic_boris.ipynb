```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
import pandas as pd

# select number of blobs and their sizes
n_samples = [2**20, 2**20]

# top-left and bottom-right centers
centers = [
    [-22,  22],   # top-left
    [ 22, -22],   # bottom-right
]

X, y = make_blobs(
    n_samples=n_samples,
    centers=centers,
    cluster_std=17,   # a bit less overlap
    random_state=42
)

print(f"Total points: {X.shape[0]:,}")
print(f"Dimensions: {X.shape[1]}")
for lbl, cnt in zip(*np.unique(y, return_counts=True)):
    print(f"Cluster {lbl}: {cnt:,} points")
```

```python
# Visualise
subset_idx = np.random.choice(X.shape[0], size=10000, replace=False)
Xs = X[subset_idx]
ys = y[subset_idx]

plt.figure(figsize=(7, 7))
plt.scatter(Xs[:, 0], Xs[:, 1], c=ys, s=3, cmap='tab10', alpha=0.5)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Synthetic 2D Dataset Gaussian blobs')
plt.xlim(-100, 100)
plt.ylim(-100, 100)
plt.grid(True, linestyle='--', alpha=0.4)
plt.tight_layout()
plt.show()
```
